```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\michael bamfo\Desktop\my corr\happyscore_income.csv")
```

```python
#a quick glimpse of the dataset
df.shape
df.info()
df.head(10)
```

```python
#dropping columnn country.1
df =df.drop(["country.1"],axis =1)

#checking for duplicates and any missibg values
df.isnull().sum()
df.duplicated().sum()
```

```python
df.head(10)
```

```python
#scatter plot with happyscore and gdp

sns.regplot(x='GDP',y= "happyScore",data =df)
```

```python
#looking at the correlation
df.corr(numeric_only=True,method ='pearson') #pearson kendal spearman
```

```python
#plotting heatmap to compare correlation amongst variables

cor_matrix = df.corr(numeric_only=True,method ='pearson') 
sns.heatmap(cor_matrix,annot =True)
plt.title("correlation matrix for numeric features")
plt.show()
```

```python
#unstack the correlation matrix to observe the correlation between the variouus columns against eachother
corr_pairs = cor_matrix.unstack()
print(corr_pairs.to_string())
```

```python
# we sort the correlation values between the various columns against each other
sorted_pairs=corr_pairs.sort_values()
print(sorted_pairs.to_string())
```

```python
# we filter column values pairs with a correlation > 0.5 between eachother
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]

high_correlation
```
